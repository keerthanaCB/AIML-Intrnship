```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\keert\Downloads\housing.csv")
```

```python
df.head()
```

```python
num_col="population"
cat_col="households"
```

```python
df.head()
```

```python
plt.figure()
sns.histplot(df[num_col], kde=True)
plt.title("Histogram with KDE")
plt.xlabel(num_col)
plt.ylabel("Frequency")
plt.show()
```

```python
skewness = df[num_col].skew()
skewness
```

```python
kurthosis = df[num_col].kurt()
kurtosis
```

```python
plt.figure()
sns.countplot(x=cat_col, data=df)
plt.title("Count plot")
plt.xticks(rotation=45)
plt.show()
```
