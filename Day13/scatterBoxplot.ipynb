```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'SquareFeet': [600, 800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],
    'Price': [15, 20, 28, 35, 50, 65, 80, 95, 120],   
    'Location': ['Urban', 'Urban', 'Semi-Urban', 'Urban', 'Semi-Urban',
                 'Rural', 'Urban', 'Urban', 'Semi-Urban']
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['SquareFeet'], df['Price'])
plt.xlabel("House Size (Square Feet)")
plt.ylabel("Price (in Lakhs)")
plt.title("Scatter Plot: House Size vs Price")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Location', y='Price', data=df)
plt.xlabel("Location")
plt.ylabel("Price (in Lakhs)")
plt.title("Box Plot: Location vs Price")
plt.show()
```

Observation:
As house size (SquareFeet) increases, the price also increases.
This shows a positive relationship between size and price.

From the box plot, Urban areas generally have higher prices
compared to Rural and Semi-Urban areas.
