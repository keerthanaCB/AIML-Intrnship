```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = {
    'Age': [22, 25, 30, 35, 40, 45, 50, 28, 33, 38],
    'Salary': [25000, 30000, 45000, 50000, 60000, 65000, 70000, 48000, 52000, 58000]
}

df = pd.DataFrame(data)
df
```

```python
scaler_standard = StandardScaler()
standardized_data = scaler_standard.fit_transform(df)

df_standardized = pd.DataFrame(
    standardized_data,
    columns=['Age_Standardized', 'Salary_Standardized']
)
```

```python
df_standardized
```

```python
scaler_minmax = MinMaxScaler()
normalized_data = scaler_minmax.fit_transform(df)

df_normalized = pd.DataFrame(
    normalized_data,
    columns=['Age_Normalized', 'Salary_Normalized']
)
```

```python
df_normalized
```

```python
plt.figure()
plt.hist(df['Salary'], bins=5)
plt.title("Original Salary")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.hist(df_standardized['Salary_Standardized'], bins=5)
plt.title("Standardized Salary")
plt.xlabel("Standardized Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.hist(df_normalized['Salary_Normalized'], bins=5)
plt.title("Normalized Salary")
plt.xlabel("Normalized Value")
plt.ylabel("Frequency")
plt.show()
```
